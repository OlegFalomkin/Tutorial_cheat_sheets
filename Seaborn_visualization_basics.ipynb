# Seaborn Basics

```python
import seaborn as sns
```

```python
import pandas as pd
```

import dataset

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

Scatter plot

```python
ax = sns.scatterplot(data=penguins, x='bill_length_mm', y='body_mass_g')
```

We can make plot using Figure-level interface for drawing relational plots onto a FacetGrid.

```python
ax = sns.relplot(data=penguins, x='bill_length_mm', y='body_mass_g', kind='scatter')
```

Use 'hue' (category within graph) and 'col' (several graphs) to divide data into needed categories

```python
ax = sns.relplot(data=penguins, x='bill_length_mm', y='body_mass_g', col='sex', hue='island')
```

'hue' can be set to numeric feature

```python
ax = sns.relplot(data=penguins, x='bill_length_mm', y='body_mass_g', hue='flipper_length_mm')
```

```python
plt.scatter(penguins['bill_length_mm'], penguins['body_mass_g'], c=[c_map[i] for i in penguins['sex']])
plt.legend(['Male','Female','nan'])
plt.show()
```

```python
legend = plt.legend()
```

```python

bill_length = penguins['bill_length_mm']
body_mass = penguins['body_mass_g']

male_data = penguins[penguins['sex'] == 'Male']
female_data = penguins[penguins['sex'] == 'Female']

scatter_male = plt.scatter(male_data['bill_length_mm'], male_data['body_mass_g'], c='blue', label='Male')
scatter_female = plt.scatter(female_data['bill_length_mm'], female_data['body_mass_g'], c='red', label='Female')


plt.legend(handles=[scatter_male, scatter_female])


plt.xlabel('Bill Length (mm)')
plt.ylabel('Body Mass (g)')
plt.title('Scatter Plot of Penguins')


plt.show()
```

displot - Figure-level interface for drawing distributions plots

```python
sns.displot(data=penguins, x='body_mass_g', kind='hist', col='species', hue='island')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='kde')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='kde', fill=True)
```

```python
sns.displot(data=penguins, x='body_mass_g',y='bill_depth_mm', kind='hist', cbar=True)
```

```python
sns.displot(data=penguins, x='body_mass_g', hue='sex', kind='hist')
```

Jointplot

```python
sns.jointplot(data=penguins, x='bill_length_mm',y='flipper_length_mm')
```

Pairplot

```python
sns.pairplot(data=penguins, hue='species')
```

Catplot

```python
sns.catplot(data=penguins, x='sex', y='body_mass_g')
```

Lmplot to plot regression
